```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, cauchy, laplace, poisson, uniform
```

```python
n = [10, 100, 1000, 10000] # Размер выборки
```

## Нормальный закон распределения

```python
def norm_distribution(n : list()):

  for i in range(len(n)):
    # Определяем количество столбцов в гистограмме в зависимости от объема выборки
    if n[i] <= 100:
      bins = 5
    else:
      bins = 50

    # формируем выборку по норм. закону распределения с E = 1, D = 1
    data = np.random.normal(0, 1, n[i])

    # Определяем E и sigma на основе сформированной выборки
    mu, std = norm.fit(data)

    # Строим гистограмму с плотностью распределения
    plt.hist(data, density=True, alpha=0.6, color='orange', bins= bins)
    plt.title(f'n = {n[i]}')

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 1000)

    # Определяем значения плотности распределения на основе определенных E и sigma из сформированной выборки
    p = norm.pdf(x, mu, std)

    # Отображение графика плотности распределения
    plt.plot(x, p, 'k', linewidth=2)
    # plt.axvline(data.mean(), color = 'k', linestyle = 'dashed', linewidth = 2)

    plt.show()

norm_distribution(n)
```

## Распределение Коши

```python
def cauchy_distribution(n : list()):
  for i in range(len(n)):
    # Определяем количество столбцов в гистограмме в зависимости от объема выборки
    if n[i] < 100:
      bins = 10
    else:
      bins = 50
    # формируем выборку по распределению Коши
    data = np.random.standard_cauchy(n[i])
    data =data[(data>-25) & (data<25)]

    # Определяем параметры на основе сформированной выборки
    x0, gam = cauchy.fit(data)

    # Построение гистграммы с плотностью распределения
    plt.hist(data, density=True, alpha=0.6, color='blue', bins = bins)
    plt.title(f'n = {n[i]}')

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 1000)

    # Определяем значения плотности распределения на основе определенных параметров из сформированной выборки
    p = cauchy.pdf(x, x0, gam)

    # Отображение графика плотности распределения
    plt.plot(x, p, 'k', linewidth=2)
    # plt.axvline(data.mean(), color = 'k', linestyle = 'dashed', linewidth = 2)

    plt.show()

cauchy_distribution(n)
```

## Распределение Лапласа

```python
def laplace_distribution(n : list()):
  mu = 0.
  lam = 2**0.5

  for i in range(len(n)):
    # Определяем количество столбцов в гистограмме в зависимости от объема выборки
    if n[i] < 100:
      bins = 10
    else:
      bins = 50
    # формируем выборку по распределению Коши

    data = np.random.laplace(mu, lam, n[i])

    # Определяем параметры на основе сформированной выборки
    a, b = laplace.fit(data)

    # Построение гистграммы с плотностью распределения
    plt.hist(data, density=True, alpha=0.6, color='blue', bins = bins)
    plt.title(f'n = {n[i]}')

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 1000)

    # Определяем значения плотности распределения на основе определенных параметров из сформированной выборки
    p = laplace.pdf(x, a, b)

    # Отображение графика плотности распределения
    plt.plot(x, p, 'k', linewidth=2)
    # plt.axvline(data.mean(), color = 'k', linestyle = 'dashed', linewidth = 2)

    plt.show()

laplace_distribution(n)
```

## Распределение Пуассона

```python
def poisson_distribution(n : list()):
  bins = 10

  for i in range(len(n)):

    # формируем выборку по распределению Коши
    lam = 10
    data = np.random.poisson(lam, n[i])

    # Определяем параметры на основе сформированной выборки
    lam_2 = np.mean(data)

    # Построение гистграммы с плотностью распределения
    plt.hist(data, density=True, alpha=0.6, color='blue', bins = bins)
    plt.title(f'n = {n[i]}')

    xmin, xmax = plt.xlim()
    x = np.arange(int(xmin), int(xmax) + 1 )

    # Определяем значения плотности распределения на основе определенных параметров из сформированной выборки
    p = poisson.pmf(x, lam_2)

    # Отображение графика плотности распределения
    plt.plot(x, p, 'k', linewidth=2)
    # plt.axvline(data.mean(), color = 'k', linestyle = 'dashed', linewidth = 2)

    plt.show()

poisson_distribution(n)
```

## Равномерное распределение

```python
def uniform_distribution(n : list()):
  low = -3**0.5
  high = 3**0.5

  for i in range(len(n)):
    # Определяем количество столбцов в гистограмме в зависимости от объема выборки
    if n[i] <= 100:
      bins = 5
    else:
      bins = 30
    # формируем выборку по распределению Коши

    data = np.random.uniform(low, high, n[i])

    # Определяем параметры на основе сформированной выборки
    a, b = uniform.fit(data)

    # Построение гистграммы с плотностью распределения
    plt.hist(data, density=True, alpha=0.6, color='blue', bins = bins)
    plt.title(f'n = {n[i]}')

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 1000)

    # Определяем значения плотности распределения на основе определенных параметров из сформированной выборки
    p = uniform.pdf(x, a, b)

    # Отображение графика плотности распределения
    plt.plot(x, p, 'k', linewidth=2)
    # plt.axvline(data.mean(), color = 'k', linestyle = 'dashed', linewidth = 2)

    plt.show()

uniform_distribution(n)
```
